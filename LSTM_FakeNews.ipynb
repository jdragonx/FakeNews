```python
import pandas as pd
import re
import numpy as np

from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('base_noticias.csv')
```

```python
df
```

```python
df = df[['Noticia','Tipo']]
df = df[df.Tipo == 0]
```

```python
df_f1
```

```python
df_f1 = df_f1[['text']]
df_f1['tipo'] = 1
```

```python
df_new = df_f1.text.str.findall(re.compile('[^\.]+')).tolist()
```

```python
df_new
```

```python
for idx in range(len(df_new)):
    if(len(df_new[idx]) > 1):
        df_new[idx] = '.'.join(df_new[idx][0:-1])
    else:
        df_new[idx] = df_new[idx][0]
        
```

```python
df_new
```

```python
df_f1.text = pd.Series(df_new)

df_f1.columns = ['Noticia','Tipo']

df = pd.concat([df,df_f1])
```

```python
df
```

```python
df_f1
```

```python
df_f1 = pd.read_csv('news_Uninverso.csv')
df_f1 = df_f1[['text']]
df_f1['tipo'] = 1

df_f1.columns = ['Noticia','Tipo']
df = pd.concat([df,df_f1])
```

```python
df
```

```python
df_train, df_test = train_test_split(df,train_size=0.8,random_state=5)
```

```python
df_test
```

```python
import tensorflow as tf
#tf.config.set_visible_devices([], 'GPU')

from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
tokenizer = Tokenizer(filters='!"#$%&()*+,-./:;<=>?@[\]^_`{|}~', lower=True)
tokenizer.fit_on_texts(df_train.Noticia.str.lower().values)
```

```python
tokenizer.word_index
```

```python
x_train = tokenizer.texts_to_sequences(df_train.Noticia.str.lower().values)
x_test = tokenizer.texts_to_sequences(df_test.Noticia.str.lower().values)
x_train = pad_sequences(x_train)
x_test = pad_sequences(x_test,maxlen=x_train.shape[1])
```

```python
x_test
```

```python
from tensorflow.keras import layers
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
np.random.seed(3)
tf.random.set_seed(3)

entrada = layers.Input(shape=x_train.shape[1])
x = layers.Embedding(len(tokenizer.word_index)+1,100,name='embedding_recurrent')(entrada)
x1 = layers.Bidirectional(layers.GRU(50, dropout=0.2, recurrent_dropout=0.1))(x)

x2 = layers.Embedding(len(tokenizer.word_index)+1,100,name='embedding_forward')(entrada)
x2 = layers.Flatten()(x2)
x2 = layers.Dense(50,activation='relu')(x2)
x2 = layers.Dropout(0.2)(x2)
x2 = layers.Dense(30,activation='softplus')(x2)
x2 = layers.Dropout(0.2)(x2)
x2 = layers.Dense(20,activation='tanh')(x2)
x2 = layers.Dropout(0.2)(x2)
x = layers.Concatenate()([x1,x2])
x = layers.Dense(50,activation='relu')(x)
x = layers.Dropout(0.2)(x)
x = layers.Dense(30,activation='softplus')(x)
x = layers.Dropout(0.2)(x)
x = layers.Dense(20,activation='tanh')(x)
x = layers.Dropout(0.2)(x)
x = layers.Dense(10,activation='exponential')(x)
x = layers.Dropout(0.2)(x)
salida = layers.Dense(1,activation='sigmoid')(x)
```

```python
model = keras.Model(
    inputs=entrada,
    outputs=salida)
model.compile(loss='msle', optimizer='nadam', metrics=['acc'])
```

```python
df_train.Tipo.value_counts()[1]
```

```python
callbacks = [
    EarlyStopping(monitor='val_acc', patience=10, min_delta=0),
    ModelCheckpoint('lstm_model.h5',monitor="val_acc",save_best_only=True)
]

weights = {
    0: df_train.Tipo.value_counts()[1],
    1: df_train.Tipo.value_counts()[0]
}

history = model.fit(x_train,df_train.Tipo,epochs=1000,validation_data=(x_test,df_test.Tipo),callbacks=callbacks,
                    batch_size=128,class_weight=weights)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
model.evaluate(x_test,df_test.Tipo)
```
